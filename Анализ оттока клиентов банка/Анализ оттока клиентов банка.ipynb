### Переименование столбцов и значений

Приведем названия столбцов к нижнему регистру и переименум столбец `'CreditCard'`:

```python
df.columns = df.columns.str.lower()
df = df.rename(columns={"creditcard": "credit_card"})
df.columns.to_list()
```

Названия столбцов успешно изменены

Проверим данные в столбцах `'City'` и `'Gender'` на уникальность:

```python
print(df['city'].value_counts())

df['gender'].value_counts()
```

Данные уникальны. Переведем их в численные значения для удобства, создав новую таблицу:

В столбце `'city'`
- `'Ярославль'` = 0
- `'Рыбинск'` = 1
- `'Ростов Великий'` = 2

В столбце `'gender'`
- `'Ж'` = 0
- `'М'` = 1

```python
df_new = df.copy()
df_new['gender'] = df_new['gender'].map({'М': 1, 'Ж': 0})
df_new['city'] = df_new['city'].map({
    'Ярославль' : 0, 
    'Рыбинск' : 1, 
    'Ростов Великий' : 2})
display(df_new.head(10))

display('Ярославль = 0, Ростов Великий = 2, Рыбинск = 1', df_new['city'].value_counts())

print("М = 1, Ж = 0\n", df_new['gender'].value_counts())
```

Визуально изменения корректны, при подсчете переименованных значений суммы совпадают с изначальными:
- Ярославль         5014
- Ростов Великий    2509
- Рыбинск           2477
- М    5457
- Ж    4543

Переименование прошло успешно.

На основании вышесказанного и имеющихся у нас данных (столбец 'credit_card') предположим, что в городах Ярославль и Рыбинск все клиенты с неизвестным балансом попали через приобретение "Кредитной карты" банка, отфильтруем данные:

```python
balance_isna.groupby(['credit_card', 'city']).agg({'score':'count'})
```

Как мы видим, предположение, что пользователи с неизвестным балансом попадают в банк исключительно через систему выдачи кредитных карт подтвердилась лишь частично - только 2592 клиента из 3617 имеют кредитную карту и это не зависит от конкретного города.

Посмотрим на соотношение числа клиентов по используемому количеству продуктов в разрезе городов:
1. общее количество пользователей
2. пользователи с положительным балансом на счёте

```python
products_group = df_new.pivot_table(index = 'products',
                                columns = 'city', 
                                values =  'userid', 
                                aggfunc = 'count')

products_group.columns = ['yaroslavl', 'rybinsk', 'rostov']
products_group['yaroslavl_ratio'] = products_group['yaroslavl'] / products_group['yaroslavl'].sum()
products_group['rybinsk_ratio'] = products_group['rybinsk'] / products_group['rybinsk'].sum()
products_group['rostov_ratio'] = products_group['rostov'] / products_group['rostov'].sum()

products_group.drop(['yaroslavl', 'rybinsk', 'rostov'], axis=1).plot(kind='bar',
                                                                     color=['#bae1ff', '#f1cbff', '#c9c9ff'])

plt.title('Распределение пользователей с любым балансом на счёте')
plt.xlabel('Количество продуктов')
plt.ylabel('Доля пользователей')
plt.xticks(rotation=0)
plt.show()
```

Как мы видели и ранее - больше всего используется 1 или 2 продукта. Также можно сказать, что:

### Построение матрицы корреляций

```python
corr = df.drop(['userid'], axis=1).corr()

plt.figure(figsize = (10, 10))

mask = np.triu(np.ones_like(corr, dtype=np.bool))

sns.heatmap(corr, 
            square = True, 
            annot = True, 
            mask=mask,
            fmt=".4f", 
            cmap='inferno',
            linewidths=.42, 
            vmax=.95)

plt.title('Матрица корреляции признаков Пирсона')
plt.show();
```

По графику можно сказать, что представленные признаки практически не коррелируют между собой. Наблюдается слабая корреляция оттока с показателями возраста и активности пользователей.

Phik ( 𝜙 k) — это новый и практичный коэффициент корреляции, который последовательно работает между категориальными, порядковыми и интервальными переменными, фиксирует нелинейную зависимость и возвращается к коэффициенту корреляции Пирсона в случае двумерного нормального входного распределения. Посмотрим:

```python
interval = list(df_new.drop(['userid'], axis=1).columns)
interval.remove('churn')
phik_overview = df_new.drop(['userid'], axis=1).phik_matrix(interval_cols=interval)

plot_correlation_matrix(phik_overview.values, 
                        x_labels=phik_overview.columns, 
                        y_labels=phik_overview.index,
                        color_map="Pastel1", 
                        title=r"Корреляция с коэффицентом $\phi$",
                        vmin=0, vmax=1,
                        fontsize_factor=1.2, 
                        figsize=(10, 8))
plt.tight_layout()
```

Корреляция с коэффицентом Phik, показывает нам слабую зависимость показателей оттока от активности клиента, умеренную от возраста клиентов и заметную для количества продуктов в использовании.

Поделим возраст на 4 группы:
- 18-31 год
- 32-44 года
- 45-60 лет
- старше 60 лет

```python
def group(age):
    if age <= 31:
        return '18-31 год'
    if age <= 44:
        return '32-44 года'
    if age <= 60:
        return '45-60 лет'
    return 'старше 60 лет'
print(group(48))
print(group(72))
print(group(26))
print(group(37))
```

Данные выводятся корректно, добавим новый столбец `age_group` с группировкой по возрастам в нашу таблицу:

`Рекомендации:` (по информации из сети интернет) в настоящее время, женщины активно интересуются (чаще чем мужчины), такими банковскими продуктами, как:

## Проверка статистических гипотез
### Гипотеза № 1

Сформулируем гипотезы:
- H0 - доход между теми клиентами, которые ушли и теми, которые остались равен
- H1 - доход между теми клиентами, которые ушли и теми, которые остались различается

1. Разделим данные на две таблицы по признаку оттока:
- churn_1 - пользователь ушел
- churn_0 - пользователь остался
2. Выведем средние значения дохода в соответствующих группах:

```python
churn_1 = df.query('churn == 1')
churn_0 = df.query('churn == 0')

print("Средний доход группы 'пользователь ушел': {:.3f}".format(
    churn_1['estimated_salary'].mean()))
print("Средний доход группы 'пользователь остался': {:.3f}".format(
    churn_0['estimated_salary'].mean()))
```

Чтобы определиться с критерием для проверки гипотезы, рассмотрим распределение признака:

```python
fig = px.histogram(df,
                   x='estimated_salary',
                   color='churn',
                   nbins=70,
                   color_discrete_sequence=['#f1cbff', '#c9c9ff'],
                   title='Распределение пользователей по доходу<br>в разрезе оттока')

fig.update_layout(bargap=0.1,
                  xaxis_title="Доход, у.е.",
                  yaxis_title="Количество пользователей, чел.")
fig.show()
```

Распределение признака не является нормальным, сравним дисперсии признака `estimated_salary` полученных выборок:

```python
var_salary_1 =  np.var(churn_1['estimated_salary'], ddof=1)
print("Дисперсия в группе 'пользователь ушел': {:.3f}".format(var_salary_1))

var_salary_0 = np.var(churn_0['estimated_salary'], ddof=1)
print("Дисперсия в группе 'пользователь остался': {:.3f}".format(var_salary_0))

print('Разница между дисперсиями:', var_salary_1 - var_salary_0)
```

Так как распределение признака не является нормальным, то мы будем использовать для проверки гипотез критерий Манна-Уитни.

Выберем критический уровень статистической значимости  0.05, как общепринятое пороговое значение.

```python
alpha = 0.05

results = st.mannwhitneyu(churn_0['estimated_salary'], churn_1['estimated_salary'])

print('p-значение: {:.3f}'.format(results.pvalue))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:**
По расчетам у нас не получилось отвергнуть нулевую гипотезу, а это значит, что скорее всего, различия дохода пользователей тех, кто ушел и тех, кто остался статистически не значимы.

### гипотеза № 2

Сформулируем гипотезы:
- H0 - средний возраст между теми пользователями, которые ушли и теми, которые остались равен
- H1 - средний возраст между теми пользователями, которые ушли и теми, которые остались различается

Выведем средние значения возраста в соответствующих группах:

```python
print("Средний возраст группы 'пользователь ушел': {:.3f}".format(
    churn_1['age'].mean()))
print("Средний возраст группы 'пользователь остался': {:.3f}".format(
    churn_0['age'].mean()))
```

На первый взгляд, средний возраст сильно отличается - на 7 лет, посмотрим, является ли это статистически значимым.

Посмотрим на распределение признака:

```python
fig = px.histogram(df,
                   x='age',
                   color='churn',
                   nbins=70,
                   color_discrete_sequence=['#f1cbff', '#c9c9ff'],
                   title='Распределение пользователей по возрасту<br>в разрезе оттока')

fig.update_layout(bargap=0.1,
                  xaxis_title="Возраст, лет.",
                  yaxis_title="Количество пользователей, чел.")
fig.show()
```

Визуально, распределение признаков близко к нормальному, однако это необходимо дополнительно проверить.

Сравним дисперсии в каждой группе:

```python
var_age_1 =  np.var(churn_1['age'], ddof=1)
print("Дисперсия в группе 'пользователь ушел': {:.3f}".format(var_age_1))

var_age_0 = np.var(churn_0['age'], ddof=1)
print("Дисперсия в группе 'пользователь остался': {:.3f}".format(var_age_0))

print('Разница между дисперсиями: {:.3f}'.format(var_age_0 - var_age_1))
```

Теперь проверим распределения нашего признака на нормальность, используем статистический тест:
1. для группы пользователей, которые ушли
2. для группы пользователей, которые остались

```python
stat, p = normaltest(churn_1['age'])
print('Statistics=%.3f, p-value=%.3f' % (stat, p))

alpha = 0.05

if p >= alpha:
    print('Данные нормально распределены')
else:
    print('Данные не распределены нормально')
```

```python
stat, p = normaltest(churn_0['age'])
print('Statistics=%.3f, p-value=%.3f' % (stat, p))

alpha = 0.05

if p >= alpha:
    print('Данные нормально распределены')
else:
    print('Данные не распределены нормально')
```

Мы получили, что:
- для группы пользователей, которые ушли - данные нормально распределены
- для группы пользователей, которые остались - данные не распределены нормально

Проведем еще один тест на нормальность - тест Шапиро-Вилка:

```python
stat, p = shapiro(churn_0['age'])
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
    print('Данные нормально распределены')
else:
    print('Данные не распределены нормально')
```

Для группы пользователей, которые остались - данные не распределены нормально.

Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, используем тест Стьюдента. Перед этим проведем тест Левена и посмотрим различаются ли дисперсии выборок, с тем, чтобы определиться какой параметр `equal_var` использовать при проведении ttest-а.

Выберем критический уровень статистической значимости  0.05, как общепринятое пороговое значение.

Сформулируем гипотезы для теста Левена:
- Н0 - дисперсии выборок равнозначны
- Н1 - дисперсии выборок различаются

```python
sample_1 = churn_0['age'].tolist()
sample_2 = churn_1['age'].tolist()

alpha = .05
stat, p = st.levene(sample_1, sample_2)
result_levene = p

if result_levene < alpha:
    print('Отвергаем Н0, дисперсии различаются, equal_var=False для ttest')
else:
    print("Не отвергаем Н0, дисперсии равнозначны, equal_var=True для ttest")
```

Используем тест Стьюдента с параметром `equal_var=False`:

```python
alpha = .05 
results = st.ttest_ind(sample_1, sample_2, equal_var=False)

print('p-значение: {:.3f}'.format(results.pvalue))

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод:**
По расчетам мы отвергли нулевую гипотезу, а это значит, что скорее всего, различия среднего возраста пользователей тех, кто ушел и тех, кто остался - 7 лет - статистически значимы.
